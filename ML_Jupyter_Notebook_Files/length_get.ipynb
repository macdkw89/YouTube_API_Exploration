```python
import pandas as pd
from googleapiclient.discovery import build
from config import api_key
import numpy as np
import isodate
```

```python
youtube = build('youtube', 'v3', developerKey=api_key)
```

```python
# load in video_df.csv
video_df = pd.read_csv('Database/ready_for_sql/video_data.csv')
```

```python
video_list = video_df['video_id'].tolist()
```

```python
# get video length
def get_video_length(video_id):
    request = youtube.videos().list(
        part='contentDetails',
        id=video_id
    )
    response = request.execute()
    return response['items'][0]['contentDetails']['duration']
```

```python
test = video_list[500]
test
```

```python
testl = get_video_length(test)

testl
```

```python
dur = isodate.parse_duration(testl).total_seconds()
```

```python
dur
```

```python
# get video length for all videos
video_length = []
for video in video_list:
    try:
        video_length.append(get_video_length(video))
    except:
        video_length.append(np.nan)
video_length
```

```python
video_df['new_length'] = video_length

video_df
```

```python
new_df = video_df
```

```python
video_df.describe()
```

```python
# convert to seconds
for i in range(len(new_df)):
    try:
        new_df['new_length'][i] = isodate.parse_duration(new_df['new_length'][i]).total_seconds()
    except:
        new_df['new_length'][i] = np.nan
```

```python
new_df
```

```python
# convert to int
new_df['new_length'] = new_df['new_length'].astype(float)
```
